```python
import ast
```

```python
def mean_reciprocal_rank(df, k=10):
    def reciprocal_rank(row):
        # Extract only the patent IDs from the tuples
        top_k_similar = [int(patent[0]) for patent in row['similar_patents'][:k]]  
        cited_patents = set(ast.literal_eval(row['cited_docdb_family_id']))  # Convert cited patents to a set for fast lookup
        
        # Find the first relevant patent in the top k
        rank = next((i + 1 for i, patent in enumerate(top_k_similar) if patent in cited_patents), None)

        # Return reciprocal rank (or 0 if no cited patent found in top k)
        return 1 / rank if rank else 0

    return df.apply(reciprocal_rank, axis=1)
```

```python
data["MRR@10"]=mean_reciprocal_rank(data)
```

```python
def mean_average_precision(df, k=1000):
    def average_precision(row):
        # Extract only the patent IDs from the tuples
        top_k_similar = [int(patent[0]) for patent in row['similar_patents'][:k]]  
        cited_patents = set(ast.literal_eval(row['cited_docdb_family_id']))  # Convert cited patents to a set for fast lookup

        relevant_found = 0  # To count how many relevant patents are found
        precision_at_k = []  # To store precision values at different ranks

        # Iterate through the top-k patents
        for i, patent in enumerate(top_k_similar):
            if patent in cited_patents:  # If the patent is relevant
                relevant_found += 1
                precision_at_k.append(relevant_found / (i + 1))  # Precision@i

        # Average Precision (AP) is the mean of Precision@i for relevant patents
        if precision_at_k:
            return sum(precision_at_k) / len(precision_at_k)
        else:
            return 0  # If no relevant patents are found in top k, AP is 0

    # Apply average_precision function to each row and calculate the mean of all Average Precision values
    return df.apply(average_precision, axis=1)

# Example Data
data["MAP@1000"]=mean_average_precision(data)
```

```python
def first_rank_of_first_relevant(df, k=1000):
    def rfr(row):
        # Extract only the patent IDs from the tuples
        top_k_similar = [int(patent[0]) for patent in row['similar_patents'][:k]]  
        cited_patents = set(ast.literal_eval(row['cited_docdb_family_id']))  # Convert cited patents to a set for fast lookup

        # Find the rank of the first relevant patent in the top-k list
        for i, patent in enumerate(top_k_similar):
            if patent in cited_patents:  # If the patent is relevant
                return i + 1  # Return the rank (1-based indexing)

        return k+1  # No relevant patent found in top k

    # Apply rfr function to each row and return the first rank of the first relevant patent
    return df.apply(rfr, axis=1)

data["RFR@1000"]=first_rank_of_first_relevant(data)
```
