```python
#loading data with representative patents and their respective keys
data_rep_patents=pd.read_csv("7similar_patents_less_than_date.csv")
```

```python
import mysql.connector
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
db_config = {
    'host': os.getenv('DB_HOST'),
    'user': os.getenv('DB_USER'),
    'password': os.getenv('DB_PASSWORD'),
    'database': os.getenv('DB_DATABASE')
}
```

```python
data_rep_patents['key']
```

```python
#sql query to get appln_ids of representative patents with the help of keys
conn = mysql.connector.connect(**db_config)
cursor = conn.cursor(dictionary=True)
keys=tuple(data_rep_patents['key'])
query = f"""
SELECT 
	t1.publn_auth,
	t1.appln_id,   
    t1.publn_nr,
    t1.publn_kind
FROM
	patstat_s24.tls211_pat_publn t1
WHERE
    CONCAT(t1.publn_auth, t1.publn_nr, t1.publn_kind) IN {keys}
"""
cursor.execute(query)
results = cursor.fetchall()
cursor.close()
```

```python
results
```

```python
#storing representative patents appln id in a dataframe
data_rep_patents_applnid=pd.DataFrame(results)
```

```python
data_rep_patents_applnid.head()
```

```python
data_rep_patents_applnid['key']=data_rep_patents_applnid['publn_auth']+data_rep_patents_applnid['publn_nr']+data_rep_patents_applnid['publn_kind']
```

```python
#performing inner join to get technology class of representative patents
data_merged=pd.merge(data_techclass, data_rep_patents_applnid, on='appln_id', how='inner')
```
