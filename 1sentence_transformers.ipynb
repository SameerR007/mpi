```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('mpi-inno-comp/paecter')
```

```python
len(model.encode(df.iloc[0]['query']['abstract']))
```

```python
df.iloc[0]['query']['title']
```

```python
import numpy as np
import torch
df['avg_sim_pos']=np.NAN
df['avg_sim_neg']=np.NAN
```

```python
#taking query as a focal and taking its respective positive documents and negative documents.
#calculating similarity between the focal embedding and each of the positive embeddings and store the average as the average positive embedding
#calculating similarity between the focal embedding and each of the negative embeddings and store the average as the average negative embedding 

for i in range(len(df)):
  print(i)
  focal=df.iloc[i]['query']['title']+" "+df.iloc[i]['query']['abstract']
  emb_focal=model.encode(focal)
  sum_sim_pos=0
  for j in range(len(df.iloc[i]['pos'])):
    pos=df.iloc[i]['pos'][j]['title']+" "+df.iloc[i]['pos'][j]['abstract']
    emb_pos=model.encode(pos)
    sim_pos=model.similarity(emb_focal,emb_pos)
    sum_sim_pos=sum_sim_pos+sim_pos
  #print(sum_sim_pos)
  df.at[i,'avg_sim_pos']=(sum_sim_pos/torch.tensor(len(df.iloc[i]['pos']))).item()
  sum_sim_neg=0
  for k in range(len(df.iloc[i]['neg'])):
    neg=df.iloc[i]['neg'][k]['title']+" "+df.iloc[i]['neg'][k]['abstract']
    emb_neg=model.encode(neg)
    sim_neg=model.similarity(emb_focal,emb_neg)
    sum_sim_neg=sum_sim_neg+sim_neg
  #print(sum_sim_neg)
  df.at[i,'avg_sim_neg']=(sum_sim_neg/torch.tensor(len(df.iloc[i]['neg']))).item()
```

```python
df.head()
```

```python
df['Result']=np.nan
```

```python
#store True in results column if average positive similarities for a query is greater than its corresponding average negative similarities
df['Result']=df['avg_sim_pos']>df['avg_sim_neg']
```
